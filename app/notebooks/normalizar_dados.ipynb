```python
import pandas as pd

mortes_ba =  pd.read_csv("../../data/raw/mortes_violentas_estado.csv", sep=",", encoding="latin-1")
```

```python
print(mortes_ba.info())
```

```python
print(mortes_ba.columns.tolist())
```

```python
feminicidio_ba = mortes_ba.loc[mortes_ba['GR_NATUREZA'] == 'FEMINICIDIO']
feminicidio_ba.rename(columns={'GR_NATUREZA': 'categoria',
                          'ANO': 'ano',
                          'MES': 'mes',
                          'QT_VITIMAS': 'vitimas',
                          'MUNICIPIO': 'municipio',
                          'REGIAO': 'regiao'
                          }, inplace=True)


```

```python
feminicidio_ba_normalizado = feminicidio_ba[['categoria', 'ano', 'mes', 'vitimas', 'municipio', 'regiao']]
feminicidio_ba_normalizado.to_csv("../../data/processed/feminicidio_ba.csv", index=False)
```
