```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import os
```

```python
dt_bahia = pd.read_csv("../../data/raw/mortes_violentas_estado.csv", sep=",", encoding="latin-1")
dt_bahia.head()
```

```python
df_2025 = dt_bahia[dt_bahia["ANO"] == 2025]
df_2025.head()
```

# Estilos

```python
pallette_global = ["#265C4B", "#146551", "#007566", "#589A8D", "#8FC1B5"]
pallette_global_v2 = [
    "#051F19", # 1. Ebony Green (Extremo Escuro)
    "#0B2D24", # 2. Deep Forest
    "#1A453B", # 3. Dark Moss
    "#265C4B", # 4. (Original 1)
    "#146551", # 5. (Original 2)
    "#007566", # 6. (Original 3)
    "#2D897A", # 7. Teal Mid
    "#589A8D", # 8. (Original 4)
    "#74ADA1", # 9. Sage
    "#8FC1B5", # 10. (Original 5)
    "#A8D0C9", # 11. Soft Mint
    "#C8E6E0"  # 12. Ice Mint (Extremo Claro)
]
test = sns.color_palette("dark")
theme_test = {
    "axes.facecolor": "#f0fdf4",      # Cor de fundo do gráfico (verde bem clarinho)
    "figure.facecolor": "#ffffff",    # Cor de fundo da imagem total
    "grid.color": "#d1d5db",          # Cor das linhas de grade
    "font.family": "sans-serif",}
```

```python
with sns.axes_style("whitegrid", rc=theme_test):
    g1 = sns.relplot(x="MES", y="QT_VITIMAS", data=df_2025, kind="line")

    g1.set_axis_labels("Mês", "Quantidade de Vítimas")
    g1.set_titles("Quantidade de Vítimas por Mês em 2025", size=16)
    g1.fig.suptitle("Análise de Mortes Violentas no Estado da Bahia - 2025", y=1.08, fontsize=14)


plt.show()
```

```python
g = sns.catplot(x="MES", y="QT_VITIMAS", data=df_2025, kind="bar", hue="MES", errorbar=None, palette=pallette_global_v2)
g.set_axis_labels("Periodo ( Mês )", "Quantidade de Vítimas")

g.set_titles("Quantidade de Vítimas por Mês em 2025", size=16)
g.fig.suptitle("Análise de Mortes Violentas no Estado da Bahia - 2025", y=1.08, fontsize=14)
```
